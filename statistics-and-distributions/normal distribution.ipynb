### Normal Distribution

Using the numpy.random.normal() method - generate a set of random numbers following normal distribution

the method takes in three key parameters..

loc: specifies the mean or center of the distribution, the location of the peak of the bell curve .

scale: determines the spread or the standard deviation of the distribution controlling how falt or narrow the graph is..

size: defines the shape of the array



```python
import numpy as np
import matplotlib.pyplot as plt
# import scipy.stats as norm
from scipy.stats import norm
```

```python
# Visualising the Normal distribution
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters for the normal distribution
loc = 0      # mean
scale = 1    # standard deviation

data = np.random.normal(loc=0, scale=1, size=1000)

# Create x values spanning the data range
x = np.linspace(min(data), max(data), 100)

plt.hist(data, bins=30, edgecolor='black', density=True)
# pdf = norm.pdf(x, loc=loc, scale=scale)
# plt.plot(x, pdf, color='red', label='Theoretical PDF')
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

The histogram represents the frequency of the generated numbers and the curve shows the theoretical pattern for comparison. The curve of a Normal Distribution is also known as the Bell Curve because of the bell-shaped curve.
