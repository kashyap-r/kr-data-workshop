## Binomial Distribution

Is the fundamental concept in Probability and Statistics

- it models the number of successes ina fixed number of independent trials where each trial has only two possible outcomes: success or failure
- this is widely used in coin flips, quality control and surveys
- uses the method numpy.random.binomial()
- It has 3 key parameters

        - n: number of trials
        - p: probability of success in each trial
        - size: the shape of the returned array

```python
# Generate the single random number 

import numpy as np 
```

```python
random_number = np.random.binomial(n=10, p=0.5)
print (random_number)
```
