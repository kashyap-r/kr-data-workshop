```python
# Pandas Series

import pandas as pd
import numpy as np

ser = pd.Series()
print ("Pandas Series: ", ser)

data = np.array(['g', 'e', 'f', 'k', 'h'])
ser = pd.Series(data)
print (ser)
```

## Pandas - Dataframes

```python
import pandas as pd

df = pd.DataFrame()
print (df)
print (type(df))
print (df.shape)
print (df.head())
print (df.columns)
# cannnot run the below on an empty datframe
# print (df.describe())
print (df.dtypes)

lst = ["my", "name", "is", "kashyap", ",", "I", "live", "in", "Bangalore"]
print (lst)
print (" ".join(lst))
# convert a list to a dataframe
df = pd.DataFrame(lst)
print (df)
```

```python
# 2 creating dataframe from dict of ndarray / lists

import pandas as pd

data = {'Name': ['James', 'Tom', 'John'],
        'Age': [20, 22, 22]}

df = pd.DataFrame(data)
print (df)
type (df)
```

```python
# Column Selection
import pandas as pd

data = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],
        'Address':['Delhi', 'Kanpur', 'Allahabad', 'Kannauj'],
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']}

df = pd.DataFrame(data)
print(df[['Name', 'Qualification']])
print (df.head())
```
