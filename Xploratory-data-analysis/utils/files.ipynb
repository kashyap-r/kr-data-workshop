```python
# Open a file
f = open("example.txt", "r")   # modes: r, w, a, rb, wb, etc.

# Read operations
content = f.read()             # read entire file
line = f.readline()            # read one line
lines = f.readlines()          # read all lines into list

# Write operations
f = open("example.txt", "w")   # overwrite if file exists
f.write("Hello, Python!\n")

# Append
f = open("example.txt", "a")   # add content to the end
f.write("Appending new line...\n")

# Insert in-between (not direct!)
with open("example.txt", "r") as f:
    lines = f.readlines()

lines.insert(1, "Inserted line here\n")  # insert after first line

with open("example.txt", "w") as f:
    f.writelines(lines)

# Always close
f.close()
```
