```python
import pandas as pd

# replace 'nifty.csv' with your downloaded CSV file path
path = 'D:/My DataSets/historicalIndexData/'

df = pd.read_csv(path+'NIFTY 100-01-05-2025-to-01-05-2026.csv', parse_dates=['Date'])
df = df.sort_values('Date').reset_index(drop=True)

# ensure column name is 'Close' (adjust if different)
df['prev_close'] = df['Close'].shift(1)
df = df.dropna(subset=['prev_close'])

positive_days = (df['Close'] > df['prev_close']).sum()
total_trading_days = len(df)

print(f"Total trading days in period: {total_trading_days}")
print(f"Positive days (Close > previous Close): {positive_days}")
```
